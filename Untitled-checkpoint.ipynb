```python

df = pd.read_csv("Walmart_Sales1.csv")


print(df)
```

```python
# Analyze the performance of sales and revenue at the city and branch level
import pandas as pd
# claculate the total slaes at city and branch
sales_performance = df.groupby (["City","Branch"])[["Quantity"]].sum().reset_index()
print ("total sale performance by city and branch")
print(sales_performance)

# Calculate the total revenue at the city and branch level
city_revenue = df.groupby("City")[["Unit price", "Quantity"]].sum().reset_index()
city_revenue["Total Revenue"] = city_revenue["Unit price"] * city_revenue["Quantity"]

branch_revenue = df.groupby(["City", "Branch"])[["Unit price", "Quantity"]].sum().reset_index()
branch_revenue["Total Revenue"] = branch_revenue["Unit price"] * branch_revenue["Quantity"]

# Print the results
print("Total Revenue by City:")
print(city_revenue[["City", "Total Revenue"]])
print("\nTotal Revenue by City and Branch:")
print(branch_revenue[["City", "Branch", "Total Revenue"]])
```

```python
# What is the average price of an item sold at each branch of the city ?
avg_price = df.groupby(["City","Branch"])["Unit price"].mean().reset_index()
avg_price["average price"] = avg_price["Unit price"]
print("average price of an item sold at each branch of the city:")
print(avg_price[["City","Branch","average price"]])
```

```python
# Analyze the performance of sales and revenue, Month over Month across the Product
# line, Gender, and Payment Method, and identify the focus areas to get better sales for
# April 2019
df = pd.read_csv("Walmart_Sales1.csv", parse_dates=['Date'])
df.dtypes
df["Date"] = pd.to_datetime(df["Date"], dayfirst= True )
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
# print(df)
monthly_sales_revenue = df.groupby(['Month', 'Product line', 'Gender', 'Payment',]).agg({'Quantity': 'sum', 'Unit price': 'sum','Unit price': 'sum'}).reset_index()

pd.set_option("display.max_rows",108)
print (monthly_sales_revenue)
```
